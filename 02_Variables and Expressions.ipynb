###PHYTON EXPRESSION

```python
"""computes the total cost of leasing a car given the down payment,
monthly rate, and number of months"""

down_payment = int(input('Enter down payment: ' ))
print(down_payment)

payment_per_month = int(input( 'Enter monthly payment: '))
print(payment_per_month)
num_months = int(input('Enter number of months: '))
print(num_months)
total_cost = down_payment + (payment_per_month * num_months)

print ('Total cost:', total_cost)
```

```python
# Lets take a look at the digits of a number. We will use the number 927 as an example.
user_val = 927
ones_digit = user_val % 10
#Ex: 927% 10 is 7
tmp_val = user_val // 10
tens_digit = tmp_val % 10
# Ex: tmp_val = 927//10 is 92%10 is 2.
tmp_val = tmp_val // 10
hundreds_digit = tmp_val % 10 # Ex: tmp_val=92//10=9, Then 9%10 is 9
print('The ones digit is:', ones_digit)
print('The tens digit is:', tens_digit)
print('The hundreds digit is:', hundreds_digit)
```
